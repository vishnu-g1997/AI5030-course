```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
meanX = 2
meanY = 3
mean = np.asarray([meanX,meanY])
varianceX = varianceY = 0.9
covXY = np.asarray([[varianceX,0.2],
         [0.2,varianceY]])
np.random.seed(10)
xyVariables = np.random.multivariate_normal ( mean, covXY, 20 )

print(f"Covariance computed by numpy with random data: \n {np.cov(xyVariables.T)} \n")
print(f"Ideal covariance: \n {covXY}) \n")

xx = np.linspace(-9,10,20)

plt.title("Independent distribtion of X, Y")
plt.scatter(xx,(xyVariables[0:,0]))
plt.scatter(xx,(xyVariables[0:,1]))
plt.show()

plt.title("Joint distribtion of X, Y")
plt.scatter((xyVariables[0:,0]),(xyVariables[0:,1]))
plt.show()
```

```python
##########################
# Finding covariance matrix of X, X-Y
# T = [[1  0 ],
#      [1 -1]]
##########################
T = np.asarray([[1 , 0 ],
     [1, -1]])
newxy = np.zeros((20,2))

covXXY = np.matmul(T,covXY)
covXXY = np.matmul(covXXY,T.T)
newxy[0:,0] = xyVariables[0:,0]
newxy[0:,1] = xyVariables[0:,0]-xyVariables[0:,1]

print(f"Covariance computed by numpy with random data: \n {np.cov(newxy.T)}")
print(f"Ideal covariance: \n {covXXY})")

xx = np.linspace(-9,10,20)
plt.title("Independent distribtion of X, X-Y")
plt.scatter(xx,(xyVariables[0:,0]))
plt.scatter(xx,(xyVariables[0:,0]-xyVariables[0:,1]))
plt.show()
plt.title("Joint distribtion of X, X-Y")
plt.scatter((xyVariables[0:,0]), (xyVariables[0:,0]-xyVariables[0:,1]))
plt.show()
```

```python
##########################
# Finding covariance matrix of X+Y, Y
# T = [[1  1 ],
#      [0  1]]
##########################
T = np.asarray([[1 , 1 ],
     [0, 1]])

newxy = np.zeros((20,2))
covXYX = np.matmul(T,covXY)
covXYX = np.matmul(covXYX,T.T)
newxy[0:,0] = xyVariables[0:,0]+xyVariables[0:,1]
newxy[0:,1] = xyVariables[0:,1]

print(f"Covariance computed by numpy with random data: \n {np.cov(newxy.T)} \n")
print(f"Covariance matrix: \n{covXYX} \n")

xx = np.linspace(-9,10,20)
plt.title("Independent distribtion of X+Y, Y")
plt.scatter(xx,(xyVariables[0:,0]+xyVariables[0:,1]))
plt.scatter(xx,(xyVariables[0:,1]))
plt.show()
plt.title("Joint distribtion of X+Y, Y")
plt.scatter((xyVariables[0:,0]+xyVariables[0:,1]),(xyVariables[0:,1]))
plt.show()
```

```python
##########################
# Finding covariance matrix of X+Y, X-Y
# T = [[1  1 ],
#      [1  -1]]
##########################
T = np.asarray([[1 , 1 ],
     [1, -1]])
covXYXY = np.matmul(T,covXY)
covXYXY = np.matmul(covXYXY,T.T)

newxy[0:,0] = xyVariables[0:,0]+xyVariables[0:,1]
newxy[0:,1] = xyVariables[0:,0]-xyVariables[0:,1]

print(f"Covariance computed by numpy with random data: \n {np.cov(newxy.T)} \n")
print(f"Idaeal Covariance matrix: \n{covXYXY} \n")

xx = np.linspace(-9,10,20)
plt.title("Independent distribtion of X+Y, X-Y")
plt.scatter(xx,(xyVariables[0:,0]+xyVariables[0:,1]))
plt.scatter(xx,(xyVariables[0:,0]-xyVariables[0:,1]))
plt.show()

plt.title("Joint distribtion of X+Y, X-Y")
plt.scatter((xyVariables[0:,0]+xyVariables[0:,1]),(xyVariables[0:,0]-xyVariables[0:,1]))
plt.show()
```
